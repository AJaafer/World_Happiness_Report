```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
warnings.filterwarnings("ignore")
print(os.listdir("C:/Users/mkay/Desktop/GitHub/WH\data"))
%matplotlib inline
```

```python
data=pd.read_csv('C:/Users/mkay/Desktop/GitHub/WH/data/2017.csv')
data.head(10)
```

```python
#setting up the dirty columns properly
data.columns=['Country', 'Happiness_Rank', 'Happiness_Score', 'Whisker_high',
       'Whisker_low', 'Economy_GDP', 'Family',
       'Health_Life_Expectancy', 'Freedom', 'Generosity',
       'Trust_Government_Corruption', 'Dystopia_Residual']
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.boxplot(data.Happiness_Score)
plt.show()
```

```python
data.corr()
```

```python
#heatmap of corellation btw columns
f,corr=plt.subplots(figsize=(13,8))
sns.heatmap(data.corr(),annot=True,linewidths=5,fmt=' .1f',ax=corr)
plt.show()
```

```python
# Line Plot
data.Freedom.plot(kind = 'line', color = 'purple',label = 'Freedom',linewidth=1,alpha = 0.5,grid = True,linestyle = ':')
data.Happiness_Score.plot(color = 'r',label = 'Score',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')
plt.legend(loc='upper right')     
plt.xlabel('x axis')              
plt.ylabel('y axis')
plt.title('Line Plot')            
plt.show()
```

```python
# Scatter Plot 
data.plot(kind='scatter', x='Happiness_Score', y='Freedom',alpha = 0.5,color = 'red')
plt.xlabel('Score')             
plt.ylabel('Freedom')
plt.title('Score-Freedom Scatter')
```

```python
# Histogram
data.Happiness_Score.plot(kind = 'hist',bins = 50,figsize = (12,12),edgecolor='black')
plt.show()
```
